```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from Excel file
df = pd.read_excel("DATA.xlsx", sheet_name="Monthly_1")

# Create subplots for each year
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 12))
axes = axes.flatten()  # Flatten to easily iterate over

for idx, year in enumerate(df.columns[1:]):  # Skip "Month" column
    ax = axes[idx]
    ax.plot(df["Month"], df[year], marker='o', linestyle='-', color='b', label=year)
    ax.set_title(f"Year {year}")
    ax.set_xlabel("Month")
    ax.set_ylabel("% Occurrence of EPB")
    ax.tick_params(axis='x', rotation=45)
    ax.legend()
    ax.grid(True)

# Hide any unused subplots
for i in range(len(df.columns[1:]), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout for better spacing
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to make space for the title
fig.suptitle("venaras EPB", fontsize=16, y=0.98) # Add title to the entire figure and adjust vertical position
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Load data from Excel file
df = pd.read_excel("DATA.xlsx", sheet_name="Monthly_1")

# Create a dropdown widget for year selection
year_options = df.columns[1:].tolist()
year_dropdown = widgets.Dropdown(options=year_options, description='Select Year:')

# Create an output widget for the plot
output = widgets.Output()

def update_plot(change):
    with output:
        clear_output(wait=True)
        selected_year = change.new

        plt.figure(figsize=(8, 6))  # Adjust figure size as needed
        plt.plot(df["Month"], df[selected_year], marker='o', linestyle='-', color='b', label=selected_year)
        plt.title(f"Year {selected_year}")
        plt.xlabel("Month")
        plt.ylabel("% Occurrence of EPB")
        plt.tick_params(axis='x', rotation=45)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()  # Adjust layout to prevent labels from overlapping
        plt.show()

year_dropdown.observe(update_plot, names='value')

display(year_dropdown)
display(output)

# Initial plot (optional)
update_plot({'new': year_options[0]})
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Load data from Excel file
df = pd.read_excel("DATA.xlsx", sheet_name="seasonal_1")

# Create a 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Convert seasons data manually as per new dataset
seasons = ["March-equinox", "June-solstice", "Sep-equinox", "Dec-solstice"]
years = list(df.columns[1:])
X, Y = np.meshgrid(range(len(years)), range(len(seasons)))
Z = df.iloc[:, 1:].values

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k')

# Label axes
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years, rotation=45)
ax.set_yticks(range(len(seasons)))
ax.set_yticklabels(seasons)
ax.set_zlabel('% Occurrence of EPB')
ax.set_xlabel('Years')
ax.set_ylabel('Seasons')
ax.set_title('3D Seasonal Occurrence of EPB')

# Rotate the view for better visualization
ax.view_init(elev=30, azim=135)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Data from the image
years = np.array([2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018])
seasons = ["March-equinox", "June-solstice", "Sep-equinox", "Dec-solstice"]
data = np.array([
    [35, 49, 37, 39, 61, 31, 59, 46, 30, 48, 44],  # March-equinox
    [27, 21, 19, 11, 13, 8, 11, 19, 13, 29, 23],  # June-solstice
    [18, 19, 32, 41, 13, 52, 17, 25, 38, 13, 18],  # Sep-equinox
    [20, 11, 12, 9, 14, 8, 13, 9, 19, 9, 14]  # Dec-solstice
])

X, Y = np.meshgrid(years, range(len(seasons)))

plt.figure(figsize=(20, 12))
cp = plt.contourf(X, Y, data, cmap='terrain', alpha= 1)
plt.colorbar(cp)

plt.yticks(range(len(seasons)), seasons)
plt.xlabel("Year")
plt.ylabel("Seasons")
plt.title("Contour Plot of Seasonal Data over Years")

plt.show()
```

```python
# prompt: separate color and shades for the year and season , also show the hover effect

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
from mpl_toolkits.mplot3d import Axes3D

# Load data from Excel file
df_seasonal = pd.read_excel("DATA.xlsx", sheet_name="seasonal_12")

# Create a 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Convert seasons data manually as per new dataset
seasons = ["March-equinox", "June-solstice", "Sep-equinox", "Dec-solstice"]
years = list(df_seasonal.columns[1:])
xpos = np.arange(len(years))
ypos = np.arange(len(seasons))
xpos, ypos = np.meshgrid(xpos, ypos)
xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros(len(xpos))
zdx = np.ones(len(xpos))
zdy = np.ones(len(ypos))
zdz = df_seasonal.iloc[:, 1:].values.flatten()

# Define a colormap
cmap = plt.cm.viridis  # You can change the colormap here

# Plot the bars with colormap and hover effect
for i in range(len(xpos)):
    color = cmap(zdz[i] / np.max(zdz))  # Normalize values for colormap
    bar = ax.bar3d(xpos[i], ypos[i], zpos[i], zdx[i], zdy[i], zdz[i], shade=True, color=color)


# Label axes
ax.set_xticks(np.arange(len(years)))
ax.set_xticklabels(years, rotation=45)
ax.set_yticks(np.arange(len(seasons)))
ax.set_yticklabels(seasons)
ax.set_zlabel('% Occurrence of EPB')
ax.set_xlabel('Years',labelpad = 15)
ax.set_ylabel('Seasons',labelpad = 15)
ax.set_title('3D Bar Chart of Seasonal Occurrence of EPB Benaras')

# Rotate the view for better visualization
ax.view_init(elev=30, azim=120)

plt.show()
```
